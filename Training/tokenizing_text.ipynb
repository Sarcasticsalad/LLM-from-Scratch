# Tokenization

```python
result = re.split(r'([,.:;?_!"()\']|--|\s)', raw_text)
```

```python
print(result)
```

```python
stripped_whitespace_text = [item.strip() for item in result if item.strip()]
print(stripped_whitespace_text)
```

```python
preprocessed_text = stripped_whitespace_text
```

```python
len(preprocessed_text)
```

# Converting the Tokens to Token IDs

```python
all_words = sorted(set(preprocessed_text))
vocab_size = len(all_words)
print(vocab_size)
```

```python
# Building the vocabulary

vocab = {token:integer for integer, token in enumerate(all_words)}

# Iterating over vocab items

for i, item in enumerate(vocab.items()):
    print(item)
    if i >= 50:
        break
```

```python
# Simple Tokenizer

class SimpleTokenizerV1:
    def __init__(self, vocab):
        # String to Integer mapping
        self.str_to_int = vocab
        # Integer to String mapping
        self.int_to_str = {i:s for s,i in vocab.items()}

    # Encoder
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        ids = [self.str_to_int[s] for s in preprocessed]

        return ids

    # Decoder
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
        
```

```python
# Initializing the tokenizer
tokenizer = SimpleTokenizerV1(vocab)
```

```python
text = """Thus hath been opened thee
This Truth of Truths, the Mystery more hid
Than any secret mystery. Meditate!
And--as thou wilt--then act!"""
```

```python
test_ids = tokenizer.encode(text)
```
